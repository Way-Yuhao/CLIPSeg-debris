```python
import os
from natsort import natsorted
```

```python
query_imgs = natsorted(os.listdir('/home/yl241/data/HIDeAI/multi_labeler_onehot/majority_vote_734_negative/original'))
len(query_imgs)
```

```python
excluded = ['001477', '000391', '000393', '020818', '020976', '000134', '021722']
filtered_imgs = [img for img in query_imgs if not any(excl in img for excl in excluded)]
len(filtered_imgs)
```

```python
dataset_ids =  [name.split('-')[2].split('_')[0] for name in filtered_imgs]
dataset_ids
```

```python
no_vis_prompts = natsorted(os.listdir('/home/yl241/data/HIDeAI/multi_labeler_onehot/majority_vote_734_negative/vis_prompts/no/'))
len(no_vis_prompts)
```

```python
no_vis_prompt_ids = [name.split('-')[2].split('_')[0] for name in no_vis_prompts]
len(no_vis_prompt_ids)
```

```python
outside_dataset = [id for id in no_vis_prompt_ids if id not in dataset_ids]
outside_dataset
```

```python
low_vis_prompts = natsorted(os.listdir('/home/yl241/data/HIDeAI/multi_labeler_onehot/clipseg_debris_dataset/vis_prompts/low/'))
low_vis_prompt_ids = [name.split('-')[2].split('_')[0] for name in low_vis_prompts]
print(len(low_vis_prompt_ids))
outside_dataset = None
outside_dataset = [id for id in low_vis_prompt_ids if id not in dataset_ids]
print(outside_dataset)
```
